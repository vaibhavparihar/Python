```python
import matplotlib.pyplot as plt

%matplotlib inline

x = list(range(0,55))
y = list(range(-55,0))

plt.plot(x,y)
```

```python
a = [0 ,3, 9 ,97 , 55 , 100]
b = [9 , 3 , 0 , 66, 75 , 4]
plt.plot(a,b)
```

```python
plt.axis([50, 110,44,90])
plt.plot(a,b)
```

```python
plt.title("Triangle")
plt.xlabel("Array A")
plt.ylabel("Array B")
plt.axis([50, 110,44,90])
plt.plot(a,b)
```
