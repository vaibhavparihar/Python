```python
import matplotlib.pyplot as plt

%matplotlib inline
```

### Scatter plot

```python
x = [1,5,7,8,3,12,17,15]
y = [5,8,3,0,14,19,12,7]

plt.scatter(x,y,c = 'red')
```

### Pie chart

```python
companies = ['Google' , 'Amazon' , 'Meta' , 'Apple']

data = [32, 25,18,25]

fig = plt.figure(figsize = (5,5))

plt.pie(data , labels = companies)
```
